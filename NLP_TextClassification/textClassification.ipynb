```python
import pandas as pd
import numpy as np
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
stop_words = stopwords.words('turkish')
```

```python
df = pd.read_csv("/content/train.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.category.value_counts()
```

```python
duplicate = df[df.duplicated("text")]
print("Duplicate Rows based on text:")
duplicate
```

```python
arr = []
for i in range(0, len(df)):
   text = str(df.text[i]).lower().split()
   non_stopwords_t = [k for k in text if k not in stop_words]
   non_stopwords_t = ' '.join(non_stopwords_t)
   arr.append(non_stopwords_t)
```

```python
df3 = pd.DataFrame(arr, columns=["nonstopwords"])
df3.head()
```

```python
import string
punc = string.punctuation
print(punc)
```

```python
def remove_punc(text):
    return text.translate(str.maketrans("","", punc))

df3["nonpunc_text"] = df3["nonstopwords"].apply(lambda text: remove_punc(text))
```

```python
df3["category"] = df["category"]
df3 = df3.drop(columns=["nonstopwords"])
df3
```

```python
df3 = df3.drop_duplicates()
df3
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("distiluse-base-multilingual-cased-v2")
```

```python
train_embeddings = model.encode(np.array(df3["nonpunc_text"]))
```

```python
df_test = pd.read_csv("/content/testText.csv")
```

```python
df_test.head()
```

```python
arr_test = []
for i in range(0, len(df_test)):
   text = str(df_test.text[i]).lower().split()
   non_stopwords_t = [k for k in text if k not in stop_words]
   non_stopwords_t = ' '.join(non_stopwords_t)
   arr_test.append(non_stopwords_t)
```

```python
df_test2 = pd.DataFrame(arr_test, columns=["nonstopwords"])
df_test2.head()
```

```python
def remove_punc(text):
    return text.translate(str.maketrans('', '', punc))

df_test2["nonpunc_text"] = df_test2["nonstopwords"].apply(lambda text: remove_punc(text))
```

```python
df_test2
```

```python
test_embeddings = model.encode(np.array(df_test2["nonpunc_text"]))
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df.category)
labels = le.transform(df3.category)
```

```python
labels
```

```python
y_true = pd.read_csv("/content/testY.csv")
```

```python
from sklearn import svm
model = svm.SVC(random_state = 2)
model = model.fit(train_embeddings, labels)
y_pred = model.predict(test_embeddings)
```

```python
y_pred = le.inverse_transform(y_pred)
```

```python
from sklearn.metrics import f1_score
"{:.2f}".format(f1_score(y_true, y_pred, average="macro"))
```
